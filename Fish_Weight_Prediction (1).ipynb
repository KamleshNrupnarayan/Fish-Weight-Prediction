# **Fish Weight Prediction**

*With a dataset of fish species, with some of it characteristic like it vertical, diagonal, length, height, and width. We will try to predict the weight of the fish based on their characteristic. We will use Linear Regression Method to see whether the weight of the fish related to their characteristic.*

```python
#import library
import pandas as pd
```

```python
#import data
fish = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Fish.csv')
```

```python
fish.head()
```

```python
fish.describe()
```

```python
fish.info()
```

```python
#define target (y) and features(x)
```

```python
fish.columns
```

```python
y = fish['Weight']
x = fish[['Category', 'Height', 'Width', 'Length1',
       'Length2', 'Length3']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
#check shape of tarin and test sample
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#train or fit model
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#predict model
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
#model accuracy
from sklearn.metrics import mean_absolute_error, r2_score
#here scoring is used in order to obtain predection that can be used as input decision
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
